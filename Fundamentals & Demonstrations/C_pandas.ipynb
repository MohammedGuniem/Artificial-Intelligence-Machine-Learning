<h1>Pandas</h1>

```python
from IPython.display import display
import pandas as pd

# Set display options to show all columns
pd.set_option('display.max_columns', None)
```

<h3>Read data from .csv file into pandas dataframe</h3>

```python
df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv')
```

<h3>View the first 5 rows of the dataset</h3>

```python
display(df.head())
```

<p>Usually our data is much too big for us to be able to display it all.</p>
<p>Looking at the first few rows is the first step to understanding our data, but then we want to look at some summary statistics.</p>
<h3>In pandas, we can use the describe method. It returns a table of statistics about the columns.</h3>
<p>We add a line in the code below to force python to display all 6 columns. Without the line, it will abbreviate the results.</p>

```python
pd.options.display.max_columns = 6
display(df.describe())
```

<strong>Note that the describe method only gives statistics for the numerical columns.</strong>

<h3>Select the column with the passenger fare.</h3>

```python
col = df['Fare']
print(col)
```

<strong>The results is a Pandas Series that is a single column from a Pandas DataFrame.</strong>

<h3>Select just the Age, Sex, and Survived columns from our original DataFrame.</h3>

```python
small_df = df[['Age',  'Sex', 'Survived']]
display(small_df.head())
```

<strong>When selecting a single column from a Pandas DataFrame, we use single square brackets. When selecting multiple columns, we use double square brackets.</strong>

<h3>Create a new column in our DataFrame that is True if the passenger is male and False if they’re female</h3>

```python
df['male'] = df['Sex'] == 'male'
display(df)
```
