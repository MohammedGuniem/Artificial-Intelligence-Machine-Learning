<h1>Overfitting in Decision Trees</h1>

<h3>Tendency to Overfit</h3>

<p><strong>Overfitting</strong> is when we do a good job of building a model for the training set, but it doesn’t perform well on the test set.</p>

<p>Decision Trees are incredibly prone to overfitting. Since they can keep having additional nodes in the tree that split on features, the model can really dig deep into the specifics of the training set.</p>

<p>Depending on the data, this might result in a model that doesn’t capture the true nature of the data and doesn’t generalize.</p>

<p>An example is if we have a single datapoint that goes to a leaf node. It might not make sense to have that additional split, as this is probably too fine-grained and is giving single datapoints from our dataset too much power, and we can see a sign of that overfitting if the leaf nodes all have few datapoints and often only one.</p>

<p>If you let a Decision Tree keep building, it may create a tree that’s overfit and doesn’t capture the essence of the data.</p>

<h3>Pruning</h3>

<p>In order to avoid overfitting, we do what’s called <strong>pruning</strong> the tree. This means we make the tree smaller with the goal of reducing overfitting.</p>

<p>There are two types of pruning: pre-pruning & post-pruning.</p>
<ul>
    <li>In pre-pruning, we have rules of when to stop building the tree, so we stop building before the tree is too big.</li>
    <li>In post-pruning we build the whole tree and then we review the tree and decide which leaves to remove to make the tree smaller.</li>
</ul>

<p>Pre-pruning techniques are easier to implement, and here is some options to consider:</p>
<ul>
    <li><strong>Max depth:</strong> Only grow the tree up to a certain depth, or height of the tree. If the max depth is 3, there will be at most 3 splits for each datapoint.</li>
    <li><strong>Leaf size:</strong>Don’t split a node if the number of samples at that node is under a threshold</li>
    <li><strong>Number of leaf nodes:</strong>Limit the total number of leaf nodes allowed in the tree</li>
</ul>

<p>
    Pruning is a balance and we can also experience <strong>Underfitting</strong> if:
</p>
<ul>
    <li>The max depth too small</li>
    <li>The leaf size is too large</li>
    <li>The number of leaf nodes too small</li>
</ul>
    
<p>This leaves us with not much of a tree and therefore no real predictive power.</p>

<strong>
    There’s no hard science as to which pre-pruning method will yield better results. In practice, we try a few different values for each parameter and cross validate to compare their performance
</strong>
