<h1>Predicting with a Decision Tree</h1>

<h3>Decision Tree Diagram</h3>

<p>Let’s look at the example Decision Tree for the Titanic dataset shown below.</p>

<p>Within each internal node:</p>
<ul>
    <li>We have the feature and threshold to split on, and</li>
    <li>The number of samples and the distribution of the target labels (# didn’t survived vs survived).</li>
</ul>

```python
from IPython.display import Image, display
display(Image(filename='./images/example_predicting_with_a_decision_tree.jpg'))
```

<p>To interpret this, let’s start by looking at the root node. It says:</p>
<strong>
    male <= 0.5<br/>
    samples = 887<br/>
    value = [545, 342]<br/>
</strong><br/>
<p>This means that the first split will be on the male column.</p>
<ul>
    <li>If the value is <=0.5 (meaning the passenger is female) we go to the left child</li>
    <li>if the value is >0.5 (meaning the passenger is male) we go to the right child</li>
</ul>
<p>We also notice:</p>
<ul>
    <li>There are 887 datapoints to start</li>
    <li>342 are positive (did survive)</li>
    <li>545 are negative cases (didn’t survive)</li>
</ul>

<p>Further down the tree if we look at the two children of the root node, we can see:</p>
<ul>
    <li>314 female passengers (Left child of root node)</li>
    <li>573 male passengers (Right child of root node)</li>
</ul>

<p>We also see that the second split for female passengers is different from the second split for male passengers.</p>

<h3>How to Make a Prediction</h3>

<p>Let’s say we’d like to use this Decision Tree to make a prediction for a passenger with these values:</p>

<ul>
    <li>Sex: female</li>
    <li>Pclass: 3</li>
    <li>Fare: 25</li>
    <li>Age: 30</li>
</ul>

<strong>
    Starting at the root node, we ask the question at each node and go to the left child if the answer is yes and to the right if the answer is no.
</strong>

<p>Below we have the question path highlighted in the decision tree to make a prediction</p>

```python
display(Image(filename='./images/hightlighted_path_decision_tree.jpg'))
```

<p>The leaf node that we end at has the following text:</p>
<strong>
27<br/>
[24, 3]
</strong>
<p>This means there are 27 datapoints in our dataset that also land at this leaf node.</p>
<p>24 of them didn’t survive and 3 of them survived.</p> 
<strong>This means our prediction is that the passenger didn’t survive.</strong>
<br/><br/>
<strong>
    Note that because there are no rules as to how the tree is developed, the decision tree asks completely different questions of a female passenger than a male passenger.
</strong>
