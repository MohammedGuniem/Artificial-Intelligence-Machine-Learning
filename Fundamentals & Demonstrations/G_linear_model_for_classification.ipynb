<h1>A Linear Model for Classification</h1>

<h3>Classification Graphically</h3>

<p>As an example we take the classic titanic dataset, and for simplicity let’s start with only two of the features (Fare and Age), since using two features enables us to visualize the data in a graph.</p>

````python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('data/titanic.csv')
````

<p>We plot that data so that on the x-axis we have the passenger’s fare and on the y-axis their age. The yellow dots are passengers that survived and the purple dots are passengers that didn’t survive.</p>

<p>We make the following observations from the data plot below:</p>
<ul>
    <li>More yellow dots at the bottom of the graph than the top, an intuitive exeplanation is because children are more likely to survive than adults</li>
    <li>More yellow dots on the right of the graph, meaning people that paid more were more likely to survive.</li>
</ul>

````python
plt.figure(figsize=(12, 6))
plt.scatter(df['Fare'], df['Age'], c=df['Survived'])
plt.xlabel('Fare')
plt.ylabel('Age')
plt.xlim(-10, 550)
plt.ylim(-10, 85)
plt.show()
````

<strong>The task of a linear model is to find the line that best separates the two classes, so that the yellow points are on one side and the purple points are on the other.</strong>

<p>An example of such a line is drawn in the plot below and is used to make predictions about new passengers in the way that if a passenger’s datapoint lies on the right side of the line, we’d predict that they survive. If on the left side, we’d predict that they didn’t survive.</p>

````python
plt.figure(figsize=(12, 6))
plt.scatter(df['Fare'], df['Age'], c=df['Survived'])
plt.plot([0, 80], [85, 5])
plt.xlabel('Fare')
plt.ylabel('Age')
plt.xlim(-10, 550)
plt.ylim(-10, 85)
plt.show()
````

<strong>The challenge of building the model will be determining what the best possible line is.</strong>

<h3>Equation for the Line</h3>

<p>A line is defined by an equation in the following form:</p>

```sh
0 = ax + by + c
```

<p>The values a, b, and c are the coefficients. Any three values will define a unique line.</p>
<p>Let’s look at a specific example of a line where the coefficients are a=1, b=-1 and c=-30.</p>

```sh
0 = (1)x + (-1)y + (-30)
```

<p>The three coefficients are: 1, -1, -30</p>

<p>Recall that we’ve been plotting our data with x axis the Fare and y axis the Age of the passenger.</p>

<p>To draw a line from an equation, we need two points that are on the line.</p>

<p>We can see, for example, that the point (30, 0) lies right on the line (Fare 30, Age 0). If we plug it into the equation, it works out.</p>

```sh
30 - 0 - 30 = 0
```

<p>We can also see that the point (50, 20) is on the line (Fare 50, Age 20).</p>

```sh
50 - 20 - 30 = 0
```

Here’s what our line looks like on the graph.

````python
plt.figure(figsize=(12, 6))  # Make the plot wider
plt.scatter(df['Fare'], df['Age'], c=df['Survived'])
plt.plot([110, 30], [80, 0])
plt.xlabel('Fare')
plt.ylabel('Age')
plt.xlim(-10, 550)
plt.ylim(-10, 85)
plt.show()
````

<strong>The coefficients of the line are what control where the line is.</strong>

<h3>Making a Prediction Based on the Line</h3>

<p>If we take a passenger’s data, we can use this equation to determine which side of the line they fall on.</p>
<p>For example, let’s say we have a passenger whose Fare is 100 and Age is 20.</p>

````python
value = (1)*100 + (-1)*20 - 30
print(value)
````

<p>Since this value is <strong>positive</strong>, the point is <strong>on the right side of the line</strong> and we’d predict that <strong>the passenger survived</strong>.</p>

<p>Now let’s say a passenger had a Fare of 10 and their Age is 50. Let’s plug these values into the equation.</p>

````python
value = (1)*10 + (-1)*50 - 30
print(value)
````

<p>Since this <strong>value is negative</strong>, the point is <strong>on the left side of the line</strong> and we’d predict that <strong>the passenger didn’t survive</strong>.</p>

<p>We can see these two points on the plot below.</p>

````python
plt.figure(figsize=(12, 6))  # Make the plot wider
plt.scatter([100, 10], [20, 50], c=[1, 0])
plt.plot([110, 30], [80, 0])
plt.xlabel('Fare')
plt.ylabel('Age')
plt.xlim(-10, 550)
plt.ylim(-10, 85)
plt.show()
````

<p>Which side of the line a point lies on determines whether we think that passenger will survive or not.</p>

<h3>What Makes a Good Line?</h3>

<p>To answer this question, let's look at 2 different lines and compare between them.</p>

<p>First we have the line we’ve been working with so far. Let’s call this Line 1.</p>

```sh
0 = (1)x + (-1)y - (30)
```

<p>Next we have another equation for a line. Let’s call this Line 2.</p>

```sh
0 = (4)x + (5)y - (400)
```

<p>We plot the 2 lines side by side along with the our data to see the difference</p>

````python
# Generate some data
import numpy as np
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and a 1x2 subplot layout
plt.figure(figsize=(11, 4))  # Adjust the figure size as needed

# Plot first line
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.scatter(df['Fare'], df['Age'], c=df['Survived'])
plt.plot([110, 30], [80, 0])
plt.title('First Line: 0 = (1)x + (-1)y - (30)')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.xlim(-10, 550)
plt.ylim(-10, 85)

plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.scatter(df['Fare'], df['Age'], c=df['Survived'])
plt.plot([180, 100], [80, 0])
plt.title('Second Line: 0 = (4)x + (5)y - (400)')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.xlim(-10, 550)
plt.ylim(-10, 85)

plt.tight_layout()  # Adjust subplot parameters to give specified padding
plt.show()
````

<p>If we look at the two lines, we see that:</p>
<ul>
    <li>Line 1 has more yellow on the right and more purple points on the left. </li>
    <li>Line 2 doesn’t have many points to the right of it; most of both the purple and yellow dots are on the left.</li>
</ul>
<strong>This makes Line 1 the preferred line as it does a better job of splitting the yellow and purple points.</strong>
