<h1>Matplotlib</h1>

<p>We can use the matplotlib library to plot our data. Plotting the data can often help us build intuition about our data.</p>

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv')
```

<h3>Scatter Plot</h3>

<h4>We use the scatter function to plot our data.</h4>
<p>The first argument of the scatter function is the x-axis (horizontal direction) and the second argument is the y-axis (vertical direction).</p>

```python
plt.scatter(df['Age'], df['Fare'], c=df['Pclass'])
plt.plot([0, 80], [85, 5])
plt.xlabel('Age')
plt.ylabel('Fare')
```

<strong>The purple dots are first class, the green dots are second class, and the yellow dots are third class.</strong>

<p>A scatter plot is used to show all the values from your data on a graph. In order to get a visual representation of our data, we have to limit our data to two features.</p>

<h3>Line Plot</h3>

```python
plt.scatter(df['Age'], df['Fare'], c=df['Pclass'])
plt.plot([0, 80], [85, 5])
plt.xlabel('Age')
plt.ylabel('Fare')
```

<strong>You can see that the yellow (3rd class) and green (2nd class) points are mostly below the line and the purple (1st class) are mostly above.</strong>
