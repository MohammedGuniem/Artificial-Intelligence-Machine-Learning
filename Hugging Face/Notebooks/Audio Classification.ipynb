<h1>Audio Classification</h1>

```python
from transformers import pipeline
from datasets import Audio, load_dataset
import librosa
from IPython.display import Audio
```

```python
dataset = load_dataset("Imama/English_Songs")

# Load the audio file using librosa
audio_data, sample_rate = librosa.load(dataset['train']['audio'][0]['path'])

# Play the audio in the notebook
Audio(audio_data, rate=sample_rate)
```

```python
classifier = pipeline(
    task="audio-classification",
    model="superb/wav2vec2-base-superb-ks",
    trust_remote_code=True
)

genreClassifier = pipeline(
    task="audio-classification",
    model="mtg-upf/discogs-maest-30s-pw-73e-ts",
    trust_remote_code=True
)
```

```python
prediction = classifier(dataset['train']['audio'][0]['array'])
print(prediction)
```

```python
prediction = genreClassifier(dataset['train']['audio'][0]['array'])
print(prediction)
```
